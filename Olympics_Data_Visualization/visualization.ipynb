```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('olympics.csv')
df.rename(columns={'Total': 'Total_Medals'}, inplace=True)
df.drop(len(df)-1, axis=0, inplace=True)
df.head()
```

# Performance of Top 10 Countries Based on Summer, Winter & Overall

```python
fig, (ax_1, ax_2, ax_3) = plt.subplots(3,1, figsize=(12,30))
plt.subplots_adjust(hspace=0.5)
df['Total_Points'] = df['Gold_Total']*3 + df['Silver_Total']*2 + df['Bronze_Total']
top_10_summer = df.nlargest(10, 'Total_Summer').set_index('Country_Name')
top_10_summer[['Gold_Summer', 'Silver_Summer', 'Bronze_Summer']].plot(kind='bar', ax=ax_1)
ax_1.set(title='Top 10 Country in Summer')
top_10_winter = df.nlargest(10, 'Total_Winter').set_index('Country_Name')
top_10_winter[['Gold_Winter', 'Silver_Winter', 'Bronze_Winter']].plot(kind='bar', ax=ax_2)
ax_2.set(title='Top 10 Country in Winter')
top_10 = df.nlargest(10, 'Total_Points').set_index('Country_Name')
top_10[['Gold_Total', 'Silver_Total', 'Bronze_Total']].plot(kind='bar', ax=ax_3)
ax_3.set(title='Top 10 Country')
```
